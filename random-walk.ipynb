# Random Walk
Random walk é um problema clássico da física estatística muito utilizado como introdução ao método estatístico. Nese notebook será desenvolvido códigos em python para apresentar suas características.
## Distribuição de Probabilidade
$$ P(n_1)=\binom{N}{n_1}p^{n_1}q^{N-n_1}$$
Onde $P(n_1)$ é a probabilidade da partícula dar $n_1$ passos a direita, $N$ é o número total de passos, $p$ e $q$ são as probabilidades de dar uma passo a direita e esqueda, respectivamente.
## Parâmetros
$$\overline{n_1} = Np$$
$$\sigma²(n_1) = Npq$$
$$\delta(n_1) = \sqrt \frac{q}{p}\frac{1}{\sqrt N}$$

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_walk(N = 40, p= 0.5):
    q=1-p
    P = np.zeros(N+1)
    n1_array = np.arange(N+1)
    with np.nditer(P, op_flags=['readwrite']) as it:
        for x, n1 in zip(it, n1_array):
            x[...] = math.comb(N,n1)*p**n1*q**(N-n1)
    plt.scatter(n1_array/N, P)
```

```python
random_walk(40,0.5)
```

```python
random_walk(400,0.5)
```

```python
random_walk(400,0.3)
```

## Distribuição Gaussiana
É dada pela seguinte distribuição:
$$p_G(N_1)=p_0\exp\left[-\frac{(N_1-\langle N_1 \rangle)²}{2\langle (N_1 - \langle N_1 \rangle)²\rangle}\right]$$
Para random walk:
$$ \langle N_1 \rangle = Np$$
$$ \langle (N_1 - \langle N_1 \rangle)²\rangle = Npq$$
Para encontrar $p_0$, usa-se a condição de normalização:
$$\lmoustache^ \infty_{- \infty}p_0\exp\left[-\frac{(N_1-\langle N_1 \rangle)²}{2\langle (N_1 - \langle N_1 \rangle)²\rangle}\right]dN_1 = 1 $$

```python
def gauss(N = 40, p= 0.5):
    q=1-p
    P = np.zeros(N+1)
    n1_array = np.arange(N+1)
    with np.nditer(P, op_flags=['readwrite']) as it:
        for x, n1 in zip(it, n1_array):
            x[...] = 1/(np.sqrt(2*np.pi*N*p*q))*np.exp(-(n1-N*p)**2/(2*N*p*q))
    plt.scatter(n1_array/N, P)
```
